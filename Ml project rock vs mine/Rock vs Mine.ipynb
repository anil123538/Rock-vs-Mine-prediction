```python
# seperating labels and data
```

```python
X=sonar_data.drop(columns=60,axis=1)  #data
Y=sonar_data[60]                      #labels
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
model=LogisticRegression() # training the model with training data
model.fit(x_train,y_train)
```

```python
# evaluting the model on training data
```

```python
x_train_prediction=model.predict(x_train)
train_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print(f"The accuracy for training data is : {train_data_accuracy}")
```

```python
# evaluating data on testing data
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print(f"The accuracy for testing data is : {test_data_accuracy}")
```

```python
print(classification_report(y_test,x_test_prediction))
```
